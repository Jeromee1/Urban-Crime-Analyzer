```python
import joblib
import osmnx as ox

tags = {'amenity': ['bar', 'atm', 'school', 'hospital']}

city_pois = {}
for city, place in [('Chicago', 'Chicago, Illinois'), 
                     ('New York', 'New York City, New York'), 
                     ('Los Angeles', 'Los Angeles, California')]:
    pois = ox.features_from_place(place, tags=tags)
    pois = pois.to_crs('EPSG:3857')
    pois['lat'] = pois.geometry.centroid.to_crs('EPSG:4326').y
    pois['lon'] = pois.geometry.centroid.to_crs('EPSG:4326').x
    pois = pois.dropna(subset=['lat', 'lon'])
    city_pois[city] = pois
```

```python
joblib.dump(city_pois, '../models/city_pois.pkl', compress=3)
```
